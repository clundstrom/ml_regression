<strong>2.1 Split</strong>

This cell splits the polynomial200.csv into two separate sets, one for training and one for testing and then
plots them side by side using using scatter-plot.


```python
import numpy as np
from os import path
import matplotlib.pyplot as plt
import kNearest
plt.rcParams["figure.figsize"] = (8,8) # scale plots 

data = np.genfromtxt(path.join('data/polynomial200.csv'), delimiter=',', dtype=float)

X_train = np.array(data[:-100])  # extract training set
X_test = np.array(data[100:])  # extract test set

# Number of sets
fig, ax = plt.subplots(nrows=1, ncols=2)

# Plot train set
ax = plt.subplot(2, 2, 1)
ax.set_title("Training set")
ax.scatter(X_train[:, 0], X_train[:, 1], s=1, color='b')

# Plot test set
ax = plt.subplot(2, 2, 2)
ax.set_title("Test set")
ax.scatter(X_test[:, 0], X_test[:, 1], s=1, color='r')
```

<strong>2.3 Plot results in 2x2</strong>

This next cell creates 4 plots for the k values [1, 3, 5, 7] and outputs the training errors for each plot.


```python
# Get min and max x-values
x_min, x_max = data[:, 0].min() - 0.1, data[:, 0].max() + 0.1  # slice array (0-118, col 0)

# Produce 100 even steps between min and max x-values
x_steps = np.linspace(x_min, x_max, num=100, dtype=np.float).T

fig, ax = plt.subplots(nrows=2, ncols=2)
plt.rcParams["figure.figsize"] = (20,20) # scale plots 
figNr = 1
for k in [1, 3, 5, 7]:
    ax = plt.subplot(2, 2, figNr)
    figNr += 1

    ax.scatter(X_train[:, 0], X_train[:, 1], s=1, color='r')
    plot = []
    # Regression line
    for x in x_steps:
        plot.append((x, kNearest.predictY(x, k, X_train)))

    plot = np.array(plot)

    # plot steps and predicted values
    ax.plot(plot[:, 0], plot[:, 1], color='b')

    y_predict = []
    for x in X_train:
        y_predict.append(kNearest.predictY(x[0], k, X_train))

    y_predict = np.array(y_predict).T
    err = round(kNearest.MSE(X_train[:, 1], y_predict[:]), 2)
    ax.set_title(f"k={k}, MSE={err}")
plt.show()
```

<strong>2.4 Test error</strong>

This next cell predicts Y-value for the test-set and computes the test error for each value of K.


```python
import numpy as np
from os import path
import kNearest

data = np.genfromtxt(path.join('data/polynomial200.csv'), delimiter=',', dtype=float)
X_test = np.array(data[100:])  # extract test set

print("Test errors: ")
for k in [1, 3, 5, 7]:
    y_test_predictions = kNearest.predictSet(k, X_test)
    err = round(kNearest.MSE(X_test[:, 1], y_test_predictions[:]), 2)

    print(f"{k}={err}")
```

<strong>2.5 MSE and K-values</strong>

I would pick 5 or 7 as the best K-value for the regression. This is because K=0 is severely overfitting
the data and yields a 0 MSE. K=3 is overfitting the data too much and the MSE rate-of-change is still high at K=3. It first starts to slow down and stabilize around K=5 and K=7 as shown in the graph below. Predictions from K=7 and up seem to underfit the model more and thus yield higher MSE values. 

A good model should have its hyperparameter chosen in or around the "elbow" of the graph such as shown in 
cell 10 where the MSE starts to stabilize a bit before increasing again.


```python
import matplotlib.pyplot as plt
import kNearest
plt.rcParams["figure.figsize"] = (8,8) # scale plots 

k_vals = []
errors = []
for k in range(1,20):
    y_test_predictions = kNearest.predictSet(k, X_test)
    err = round(kNearest.MSE(X_test[:, 1], y_test_predictions[:]), 2)
    errors.append(err)
    k_vals.append(k)

plt.title("Relationship of K and MSE graph")
plt.xlabel('K-value')
plt.ylabel('MSE')
plt.scatter(3, 17.59, marker='x' ,color='r')
plt.text(3.4, 17.59, "(3, 17.59)", fontsize=9)
plt.scatter(5, 21.44, marker='x', color='r')
plt.text(5.4, 21.44, "(5, 21.44)", fontsize=9)
plt.scatter(7, 23.2, marker='x', color='r')
plt.text(7.4, 23.2, "(5, 21.44)", fontsize=9)
plt.plot(k_vals, errors)
plt.show()
```
